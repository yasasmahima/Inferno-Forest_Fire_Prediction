```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline

dataset = pd.read_csv('dataset.csv')
```

```python
dataset.shape
```

```python
dataset.isnull().any()
```

```python
dataset=dataset.dropna(axis=1)
```

```python
dataset.columns
```

```python
dataset.describe()
```

```python
dataset = dataset.fillna(method='ffill')
```

```python
y = dataset[['temp', 'humidity','wind_speed']].values
y=pd.DataFrame(y)

X=dataset[['dt', 'timezone', 'lat', 'lon',
       'feels_like', 'temp_min', 'temp_max', 'pressure', 'wind_deg', 'clouds_all', 'weather_id', 'weather_main',
       'weather_description', 'weather_icon','temp']]
```

```python
x_correlations = X.corr()
```

```python
x_correlations
```

```python
(x_correlations
     .temp
     .drop('temp')
     .sort_values(ascending=False)
     .plot
     .barh())
```

```python
x_heat = sns.heatmap(x_correlations)
```

```python
X = pd.concat([X, pd.get_dummies(X['weather_main'],prefix='weather_main')], axis=1)
del X['weather_main']
```

```python
X = pd.concat([X, pd.get_dummies(X['weather_description'],prefix='weather_description')], axis=1)
del X['weather_description']
```

```python
X = pd.concat([X, pd.get_dummies(X['weather_icon'],prefix='weather_icon')], axis=1)
del X['weather_icon']
```

```python
X.columns
```

```python
x_correlations = X.corr()
```

```python
x_correlations
```

```python
x_heat = sns.heatmap(x_correlations)
```

```python
(x_correlations
     .temp
     .drop('temp')
     .sort_values(ascending=False)
     .plot
     .barh())
```

```python
X['temp_mean']=(X['temp_min']+X['temp_max'])/2
```

```python
del X['temp']
del X['temp_max']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
y_pred = regressor.predict(X_test)
```

```python
print("Error Calculations for Temperature Prediction")
print('\nMean Absolute Error:', metrics.mean_absolute_error(y_test[0], y_pred[:,0]))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test[0], y_pred[:,0]))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test[0], y_pred[:,0])))

print("\nError Calculations for Realtive Humidity Prediction")
print('\nMean Absolute Error:', metrics.mean_absolute_error(y_test[1], y_pred[:,1]))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test[1], y_pred[:,1]))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test[1], y_pred[:,1])))

print("\nError Calculations for Wind Speed Prediction")
print('\nMean Absolute Error:', metrics.mean_absolute_error(y_test[2], y_pred[:,2]))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test[2], y_pred[:,2]))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test[2], y_pred[:,2])))
```

```python
y_pred= np.array(y_pred)
y_pred[:,2]
```

```python
from sklearn.metrics import r2_score
print("Accuracy Rate For Temperature Prediction")
r2_score(y_test[0], y_pred[:,0])
```

```python
from sklearn.metrics import r2_score
print("Accuracy Rate For Relative Humidity Prediction")
r2_score(y_test[1], y_pred[:,1])
```
